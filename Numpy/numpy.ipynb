## Computation on Arrays: Broadcasting

```python
#Adding one-dimensional and two-dimensional arrays
# a is stretched or broadcast to across the second dimension in order to match the shape of c
c = np.ones((3, 3))
print(c)
```

```python
#the following is broadcasting of both arrays
x = np.array([1, 2, 3])
y = np.arange(3)[:, np.newaxis]
print(x)
print(y)
```

```python
#both arrays are stretched or broadcast
x + y
```

### Rules of Broadcasting

```python
Image(filename="rules_of_broadcasting.png", width = 800, height = 400)
```

```python
#Broadcasting Example I
M = np.ones((2,3))
a = np.arange(3)

"""
M.shape -> (2,3)
a.shpae -> (3,)

By rule 1:
M.shape -> (2,3)
a.shape -> (1,3)

By rule2:
M.shape -> (2,3)
a.shape -> (2,3)

The shapes matched!
"""
M + a
```

```python
#Broadcasting Example II
M = np.arange(3).reshape((3,1))
a = np.arange(3)

"""
M.shape -> (3,1)
a.shape -> (3,)

by rule 1:
M.shape -> (3,1)
a.shape -> (1,3)

By rule2:
M.shape -> (3,3)
a.shape -> (3,3)

The shapes matched!
"""
M + a
```

```python
#Broadcasting Example III
M = np.ones((3,2))
a = np.arange(3)

"""
M.shape -> (3,2)
a.shape -> (3,)

By rule 1:
M.shape -> (3,2)
a.shape -> (1, 3)

By rule 2:
M.shape -> (3,2)
a.shape -> (3,3)

The shapes not matched! The two arrays are not compatible.
"""
M + a
```

## Sorting Arrays

```python
#Selection Sort
x = np.array([2, 1, 4, 3, 5])
for i in range(len(x)):
    swap = i + np.argmin(x[i:])
    (x[i], x[swap]) = (x[swap], x[i])
#not very efficient for Very Large DataSet
x
```

```python
#Fast Sorting in Numpy: np.sort & np.argsort
```

```python
x = np.array([2,1,4,3,5])
np.sort(x)
```

```python
#also available in-place
x = np.array([2,1,4,3,5])
x.sort()   
print(x)
```

```python
#argsort return the indices of the sorted elements
x = np.array([2,1,4,3,5])
np.argsort(x)
```

## Structured Data: Numpy's Structured Arrays

```python
#a simple array
x = np.zeros(4, dtype = int)
x
```

```python
name = ['Alice', 'Bob', 'Cathy', 'Doug']
age = [25, 45, 37, 19]
weight = [55.0, 85.5, 68.0, 61.5]

#Creating empty container array
data = np.zeros(4, dtype = {'names':("name", "age", "weight"),
                            'formats':('U10', 'i4', 'f8')})      #4 is the number of rows in container
print("Empty Container: ",data.dtype, "\n")

#filling in data
data['name'] = name
data['age'] = age
data['weight'] = weight
print("Filled with data:")
print(data)
```

```python
#index Structuring array

#get all names
print("all names: ", data['name'])

#get the first row
print("The first row: ", data[0])

#get the name of last row
print("The name of last row: ", data[-1]['name'])

#Get names where age is under 30
print("names under 30: ", data[data['age'] < 30]['name'])
```

```python
np.dtype({'names':('name', 'age', 'weight'),
         'formats':((np.str_, 10), int, np.float32)})
```

```python
#a compound type can also be specified by a list of tuples
np.dtype([('name', 'S10'), ('age', 'i4'), ('weight', 'f8')])
```

```python
np.dtype('S10,i4,f8')
```

### More Advanced Compound Types

```python
#an element contain a matrix
tp = np.dtype([('id', 'i8'), ('mat', 'f8', (3,3))])
X = np.zeros(1, dtype = tp)
print("Container: ", X[0])
print(X['mat'][0])
```
